```python
import numpy as np
import pandas as pd
```

```python
# creating data-Frame
r_Movies = pd.read_csv('tmdb_5000_movies.csv')
r_Credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Renaming column "movie_id" to "id" in dataFrame to match the name with the column of movies dataset
r_Credits.rename(columns = {'movie_id':'id','title':'m_title'},inplace=True)
r_Credits.head()
```

```python
# merge two DataFrames based on the movie_id
movies = r_Movies.merge(r_Credits,on='id')
```

```python
# the shape of the dataframe is 23 because we have taken "id" to join two dataframes
movies.shape
```

```python
movies.head()
```

```python
# Here we are removing the columns which are not going to help to make tags for movies

# we will keep the columns like overview, title, keywords, id, geners, cast, crew
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.isnull().sum() # checking for missing data
```

```python
movies.dropna(inplace=True) #dropping the missing data
```

```python
movies.duplicated().sum() #checking for duplicates
```

```python
import ast
def transform(obj):
    list = []
    for i in ast.literal_eval(obj):   # for appeding in the empty list we have to convert string of list to list
        list.append(i['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(transform)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(transform)
```

```python
movies.head()
```

```python
# we want top three names of actors from the movie
import ast
def transform2(obj):
    list1 = []
    ctr = 0
    for i in ast.literal_eval(obj):   # for appeding in the empty list we have to convert string of list to list
        if ctr != 3:
            list1.append(i['name'])
            ctr+=1
        else:
            break
    return list1
```

```python
movies['cast'] = movies['cast'].apply(transform2)
```

```python
movies.head()
```

```python
# from crew column we want only director
import ast
def get_director(obj):
    list1 = []
    for i in ast.literal_eval(obj):   # for appeding in the empty list we have to convert string of list to list
        if i['job']=='Director':
            list1.append(i['name'])
            break
    return list1
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head()
```

```python
movies['overview'].apply(lambda x:x.split())
```

```python
#  to remove the spaces between two words
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
# Merging columns = overview, genre, keyword, cast, crew into the tag
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```
